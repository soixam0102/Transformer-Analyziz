# BLEU
- Một matrix score chuyên đánh giá độ chính xác của một mô hình dịch máy.
- Không giống như việc đánh giá một mô hình phân loại bình thường khi mà chỉ cần so sánh đầu ra với một grouth truth duy nhất và cố định. Trong dịch máy, một câu có thể có nhiều nghĩa khác nhau và được dịch khác nhau.
- Nếu sử dụng precision làm thước đo, chúng ta sẽ đếm các từ xuất hiện trong máy dịch chia cho độ dài của bản dịch. Điều này là không phù hợp khi mà nhiều bản dịch không tốt về nghĩa nhưng vẫn chứa các từ có chính xác thì cho percision cao. 
- Để đánh giá bản dịch tốt hơn, chúng ta cần đánh giá các từ theo bối cảnh ngữ nghĩa của nó nữa, ta có thể xét các từ cụm từ thường liên kết với nhau và xét tần số xuất hiện của nó trong bộ từ có sắn, cách naỳ có thể khắc phục được việc một từ xuất hiện nhiều lần nhưng vẫn mang precision cao như đã đề cập trước.

- Còn đối với BLEU sẽ có tác dụng đánh giá kết quả càng cao nếu kết quả của MT là sát nghĩa với kết quả của người dịch. nó được tính toán dựa trện P1,P2,P3,P4 theo lũy thừa cơ số tự nhiên e.

- Tuy nhiên hạn chế của bleu score là đối với các câu ngắn thì xu hướng blue score sẽ càng cao. Nên cần một chỉ số phạt độ ngắn là Brevity Penalty (BP)


```python
from nltk.translate.bleu_score import sentence_bleu
```

```python
# Lưu ý về chiều của reference và chiều của MT_translation là khác nhau
reference = [
    ['the', 'quick', 'brown', 'fox', 'jumped', 'over', 'the', 'lazy', 'dog'],
    ['the', 'strong', 'brown', 'fox', 'jumped', 'over', 'the', 'cute', 'cat']
]
MT_translation = ['the', 'fast', 'brown', 'fox', 'jumped', 'over', 'the', 'sleepy', 'dog']
score = sentence_bleu(reference,MT_translation)
print(score)
```
